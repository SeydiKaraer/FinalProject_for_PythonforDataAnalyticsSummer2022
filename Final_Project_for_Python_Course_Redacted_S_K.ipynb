```python
import pandas as pd #pandas for dataset read
import numpy as np #numpy for array
import matplotlib.pyplot as plt # matplotlib for graphs
import seaborn as sns #seaborn for visualize
#importing libraries
```

```python
df = pd.read_csv('https://github.com/SeydiKaraer/FinalProject_for_PythonforDataAnalyticsSummer2022/raw/main/world-happiness-report.csv') #reading dataset
df.head() # printing starting 5 rows of data
```

```python
df.info() #showing info
```

```python
df.duplicated().sum() #checking if there is duplicated values exist
```

```python
df.describe()
```

```python
cols = df.columns # assigning columns of dataset (features ) to cols
country_values = df[cols[0]].value_counts() #counting all the countries
print(f"Countries \n{country_values}")
```

```python
df[cols[1]].value_counts() #counting the years
```

```python
sns.set_theme(style="whitegrid") #setting colour theme for graph
sns.countplot( y= df[cols[1]] , label='count') # counting all the year wise data
```

```python
i = 1
for c in cols[2:]: #starting from 2nd to end of columns
    plt.subplot(3, 3, i) #making 3x3 subplots
    plt.hist(df[c]) #mkaing histogram for all columns using c
    plt.title((c)) # adding title of particular column
    i += 1
plt.tight_layout()
plt.show()
#code used for histogram
```

```python
df.iloc[:, 2:].plot.box(rot=90)# checking for outliers using boxplot
```

```python
df.isnull().sum()#checking null values in each column
```

```python
for c in cols[3:]:
    df.fillna(df[c].mean(), inplace=True)#filling all the null values to remove them
df.isnull().sum()
```

```python
df["score"]=(
    df["Life Ladder"] + 
     df["Log GDP per capita"] + 
     df["Social support"] + 
     df["Healthy life expectancy at birth"] + 
     df["Freedom to make life choices"] + 
     df["Generosity"] + 
     df["Positive affect"]
    ) - (df["Perceptions of corruption"] + 
     df["Negative affect"]
    )
    #assigning the score feature to all values accordingly to check for further analysis
```

```python
df.info()
```

```python
sns.heatmap(df.corr(), cmap='BuPu', square = True)#naming heat map to see correlation
```

```python
country_total_lifeladder = df.groupby('Country name')['Life Ladder'].mean().sort_values(ascending= False).head(10)# showing top 10 score country wise
country_total_lifeladder
```

```python
sns.barplot(x = country_total_lifeladder.values, y = country_total_lifeladder.index, orient='h')#barplot for all country score
```

```python
country_total_score = df.groupby('Country name')['score'].mean().sort_values(ascending= False).head(10)# showing top 10 score country wise
country_total_score
```

```python
sns.barplot(x = country_total_score.values, y = country_total_score.index, orient='h')#barplot for all country score
```

```python
social_country = df.groupby('Country name')['Social support'].mean().sort_values(ascending=False).head(10)
print(social_country)#showing the top 10 score country wise
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x = social_country.values, y = social_country.index, orient='h').set_title('Dependence social support on country')
plt.show()
```

```python
#also we need to see bottom-10 countries by mean life ladder values
total_country_bad_lifeladder = df.groupby('Country name')['Life Ladder'].mean().sort_values(ascending=False).tail(10)
print(total_country_bad_lifeladder)
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x = total_country_bad_lifeladder.values, y = total_country_bad_lifeladder.index, orient='h').set_title('Bottom-10 countries by life ladder value')
plt.show()
```

```python
#also we need to see bottom-10 countries by total score of happiness
total_country_bad_score = df.groupby('Country name')['score'].mean().sort_values(ascending=False).tail(10)
print(total_country_bad_score)
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x = total_country_bad_score.values, y = total_country_bad_score.index, orient='h').set_title('Bottom-10 countries by total score of happiness')
plt.show()
```

```python
# Score
sns.relplot(x='Healthy life expectancy at birth', y='score', data=df, kind='scatter').set(title='Dependence of Healthy life expectancy at birth on total score of happiness')
```

```python
# Life Ladder
sns.relplot(x='Healthy life expectancy at birth', y='Life Ladder', data=df, kind='scatter').set(title='Dependence of Healthy life expectancy at birth on life ladder')
```

```python
plt.figure(figsize=(20, 6))
sns.boxplot(x='year', y='Generosity', data=df).set(title='Generosity changing in years')
plt.show()
```

```python
sns.relplot(x='Healthy life expectancy at birth', y='Log GDP per capita', data=df, kind='scatter')
plt.show()
```
